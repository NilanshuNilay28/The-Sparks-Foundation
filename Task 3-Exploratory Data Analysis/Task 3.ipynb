## Segment distribution of discount, sales and profit

```python
segmentwise_distribution = df.groupby(['Segment'])[['Discount', 'Sales', 'Profit']].mean()
segmentwise_distribution
```

```python
colors= ('orange', 'cyan', 'magenta')
```

```python
segmentwise_distribution.plot.pie(subplots=True,
                                  figsize=(20,20),
                                  autopct='%1.1f%%',
                                  labels=segmentwise_distribution.index,
                                  startangle=90,
                                  colors=colors,
                                  shadow=True)
plt.title('Segment-wise distribution of Sales, Discount and Profit', fontsize=20)
plt.show()
```

### Analysis: From the above pie plots-
Discount distribution in respective segments are listed below:
Consumer 34%
Home Office 31.7%
Corporate 34%

### Sales distribution in respective segments are listed below:
Consumer 32%
Home Office 34.5%
Corporate 33.5%

### Profit distribution in respective segments are listed below:
Consumer 28.7%
Home Office 37.5%
Corporate 33.8%

Discount and Sales are higher in consumer segment but low in profit.

```python
statewise_distribution= df.groupby(['State'])[['Discount', 'Sales', 'Profit']].mean()
statewise_distribution.head()
```

```python
state_profit_sort= statewise_distribution.sort_values('Profit')
state_profit_sort['Profit'].plot(kind='bar', figsize=(20,8), color='red')
plt.title('State-wise Distribution of Profit', fontsize=25)
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```
